Pregunta NÂ°4.- Si consideran solo el grupo de pasajeros que pagÃ³ menos de $10, Â¿cuÃ¡l serÃ­a el porcentaje de este grupo que sobreviviÃ³?

```python
fareFilter = df.loc[:, 'Fare'] < 10
premulti = df.loc[fareFilter, 'Survived'].value_counts(normalize=True)
premulti = premulti * 100
print(premulti)
```
