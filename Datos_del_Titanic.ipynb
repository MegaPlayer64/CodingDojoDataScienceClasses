# **Ejercicios de Pandas en el conjunto de datos de Titanic (base)**

Pregunta N°1.- Muestren el centro del marco de datos.

```python
import pandas as pd
filename = '/content/titanic.csv'
df = pd.read_csv(filename)
df.head()
```

Pregunta N°2.- ¿Cuál es el porcentaje de las personas que sobrevivió? (pista: encuentren la media de la columna de supervivencia).

```python
Survived_Mean = df.loc[:, 'Survived'].mean()
print(Survived_Mean)
```

Pregunta N°3.- ¿Cuántos de cada sexo sobrevivieron?


```python
SurvivedConfirmed = df['Survived']==1
GenderSurvivedThing = df['Sex']
TotalSurvival = df.loc[SurvivedConfirmed & GenderSurvivedThing, :]
MaleGenderFilter = TotalSurvival.loc[:, 'Sex'] == "male"
FemaleGenderFilter = TotalSurvival.loc[:, 'Sex'] == "female"
TotalSurvival.head(9874234)
```

Pregunta N°4.- Si consideran solo el grupo de pasajeros que pagó menos de $10, ¿cuál sería el porcentaje de este grupo que sobrevivió?

```python
fareFilter = df.loc[:, 'Fare'] < 10
premulti = df.loc[fareFilter, 'Survived'].value_counts(normalize=True)
premulti = premulti * 100
print(premulti)
```

Pregunta N°5.- En total, ¿cuál es la edad promedio de los que no sobrevivieron?

```python
DeadConfirmed = df['Survived']==0
Age = df['Age']
TotalDead = df.loc[DeadConfirmed & Age, :]
TotalDead['Age'].mean()
```
