## **Cargar y filtrar datos (Base)**


```python
import pandas as pd
filename = '/content/RailsToTrails_National_Count_Data_week.xlsx'
df = pd.read_excel(filename)
df = df.rename(columns={'Change 2019-2020': 'change_2019_2020'})
pd.options.display.float_format = '{:,.2f}'.format
df['change_2019_2020'] = df['change_2019_2020'] * 100
data_post_july_filter = df[31:]
data_double_change_filter = df.loc[df['change_2019_2020']>=100]
dfnew = pd.concat([data_double_change_filter, data_post_july_filter])
dfnew.head(30)
```
