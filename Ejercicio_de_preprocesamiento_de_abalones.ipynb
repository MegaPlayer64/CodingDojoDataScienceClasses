# **Tarea N°0:**
**Precargar librerias y archivos:**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.impute import SimpleImputer

filename = '/content/abalone.data'
names = '/content/abalone.names'
df = pd.read_csv(filename)
df.columns = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
df.head
```

# **Tarea N°1:**
**Separar los datos en un matriz de características (X) y un vector objetivo (y):**

```python
# El objetivo que intentamos predecir.
y = df['Rings']
# Las características que se usarán para hacer la predicción
X = df.drop(columns=['Sex','Rings'])
```

# **Tarea N°2:**
**Crear un train test split en los datos. Utilicen un número aleatorio 42 por coherencia.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# **Tarea N°3:**
**Utilizar transformadores de columna para transformar las columnas adecuadas.**

A.- Utilizar selectores de columnas para seleccionar las columnas categóricas y las columnas numéricas.

B.- Utilizar un OneHotEncoder para codificar las columnas categóricas.

```python
ring_selector = make_column_selector(dtype_include='object')
train_ring_data = X_train[ring_selector(X_train)]
test_ring_data = X_test[ring_selector(X_test)]
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(train_ring_data)
train_ohe = encoder.transform(train_ring_data)
test_ohe = encoder.transform(test_ring_data)
train_ohe
```

C.- Utilizar StandardScaler para escalar columnas numéricas.

```python
print(X_train)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)
train_scaled[:5]
X_train_scaled = pd.DataFrame(train_scaled, columns=X_train.columns)
X_train_scaled.head()
```

 D.- Utilizar un ColumnTransformer para emparejar la transformación al tipo de columna.

```python
obj_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')
freq_imputer = SimpleImputer(strategy='most_frequent')
mean_imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
numeric_pipe = make_pipeline(mean_imputer, scaler)
categorical_pipe = make_pipeline(freq_imputer, ohe)
number_tuple = (numeric_pipe, num_selector)
category_tuple = (categorical_pipe, obj_selector)
preprocessor = make_column_transformer(number_tuple, category_tuple)
preprocessor
```

E.- Transformar los datos y mostrar los arrays de NumPy resultantes.

```python
preprocessor.fit(X_train)
X_train_processed = preprocessor.transform(X_train)
X_test_processed = preprocessor.transform(X_test)
print(np.isnan(X_train_processed).sum().sum(), 'missing values in training data')
print(np.isnan(X_test_processed).sum().sum(), 'missing values in testing data')
print('\n')
print('All data in X_train_processed are', X_train_processed.dtype)
print('All data in X_test_processed are', X_test_processed.dtype)
print('\n')
print('shape of data is', X_train_processed.shape)
print('\n')
X_train_processed
```
